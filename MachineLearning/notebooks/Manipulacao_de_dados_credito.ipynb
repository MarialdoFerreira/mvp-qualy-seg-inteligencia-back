#### **Tratando os dados**

```python
# removendo ponto do valor e substituindo a virgula pelo ponto
for atributo in ['limite_credito', 'valor_transacoes_12m']:
    n_atributo = data[atributo].str.replace('.', '')
    data[atributo] = n_atributo.str.replace(',', '.')
```

```python
# convertendo os atributos para numerico
for atributo in ['limite_credito', 'valor_transacoes_12m']:
    data[atributo] = pd.to_numeric(data[atributo])
```

```python
# Transformando Sexo Feminino de "F"para 0
for atributo in ['sexo']:
    data[atributo] = data[atributo].str.replace('F', '0')
```

```python
# Transformando Sexo Feminino de "M"para 1
for atributo in ['sexo']:
    data[atributo] = data[atributo].str.replace('M', '1')
```

```python
# convertendo os atributos para numerico
for atributo in ['sexo']:
    data[atributo] = pd.to_numeric(data[atributo])
```

```python
# Transformando Escolaridade 'ensino medio' para 0 
for atributo in ['escolaridade']:
    data[atributo] = data[atributo].str.replace('ensino medio', '0')
```

```python
# Transformando Escolaridade 'mestrado' para 1 
for atributo in ['escolaridade']:
    data[atributo] = data[atributo].str.replace('mestrado', '1')
```

```python
# Transformando Escolaridade 'sem educacao formal' para 2 
for atributo in ['escolaridade']:
    data[atributo] = data[atributo].str.replace('sem educacao formal', '2')
```

```python
# Transformando Escolaridade 'na' para 2 
for atributo in ['escolaridade']:
    data[atributo] = data[atributo].str.replace('na', '2')
```

```python
# Transformando Escolaridade 'graduacao' para 3 
for atributo in ['escolaridade']:
    data[atributo] = data[atributo].str.replace('graduacao', '3')
```

```python
# Transformando Escolaridade 'doutorado' para 4 
for atributo in ['escolaridade']:
    data[atributo] = data[atributo].str.replace('doutorado', '4')
```

```python
# convertendo os atributos para numerico
for atributo in ['escolaridade']:
    data[atributo] = pd.to_numeric(data[atributo])
```

```python
data['estado_civil'].unique()
```

```python
# Transformando estado_civil 'solteiro' para 0 
for atributo in ['estado_civil']:
    data[atributo] = data[atributo].str.replace('solteiro', '0')
```

```python
# Transformando estado_civil 'divorciado' para 1 
for atributo in ['estado_civil']:
    data[atributo] = data[atributo].str.replace('divorciado', '1')
```

```python
# Transformando estado_civil 'casado' para 2 
for atributo in ['estado_civil']:
    data[atributo] = data[atributo].str.replace('casado', '2')
```

```python
# Transformando estado_civil 'na' para 3 - viuvo
for atributo in ['estado_civil']:
    data[atributo] = data[atributo].str.replace('na', '3')
```

```python
# convertendo os atributos para numerico
for atributo in ['estado_civil']:
    data[atributo] = pd.to_numeric(data[atributo])
```

```python
# Transformando tipo_cartao 'blue' para 0
for atributo in ['tipo_cartao']:
    data[atributo] = data[atributo].str.replace('blue', '0')
```

```python
# Transformando tipo_cartao 'silver' para 1
for atributo in ['tipo_cartao']:
    data[atributo] = data[atributo].str.replace('silver', '1')
```

```python
# Transformando tipo_cartao 'gold' para 2
for atributo in ['tipo_cartao']:
    data[atributo] = data[atributo].str.replace('gold', '2')
```

```python
# Transformando tipo_cartao 'platinum' para 3
for atributo in ['tipo_cartao']:
    data[atributo] = data[atributo].str.replace('platinum', '3')
```

```python
# convertendo os atributos para numerico
for atributo in ['tipo_cartao']:
    data[atributo] = pd.to_numeric(data[atributo])
```

```python
# Transformando salario_anual 'menos que $40K' para 0
for atributo in ['salario_anual']:
    data[atributo] = data[atributo].str.replace('menos que $40K', '0')
```

```python
# Transformando salario_anual '$40K - $60K' para 1
for atributo in ['salario_anual']:
    data[atributo] = data[atributo].str.replace('$40K - $60K', '1')
```

```python
# Transformando salario_anual '$60K - $80K' para 2
for atributo in ['salario_anual']:
    data[atributo] = data[atributo].str.replace('$60K - $80K', '2')
```

```python
# Transformando salario_anual '$80K - $120K' para 3
for atributo in ['salario_anual']:
    data[atributo] = data[atributo].str.replace('$80K - $120K', '3')
```

```python
# Transformando salario_anual '$120K +' para 4
for atributo in ['salario_anual']:
    data[atributo] = data[atributo].str.replace('$120K +', '4')
```

```python
# Transformando salario_anual 'na' para 4 - '$120K +'
for atributo in ['salario_anual']:
    data[atributo] = data[atributo].str.replace('na', '4')
```

```python
# convertendo os atributos para numerico
for atributo in ['salario_anual']:
    data[atributo] = pd.to_numeric(data[atributo])
```

```python
# salva dataset (separado = ',')
data.to_csv('../data/credito-preprocessado.csv', index=False)

# Após esta salva o dataset foi disponibilizado no GitHub no link abaixo:
# Link: "https://raw.githubusercontent.com/MarialdoFerreira/mvp-qua-seg-inteligencia-back/refs/heads/main/credito-preprocessado.csv"
```
